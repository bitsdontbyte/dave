```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read survey data
df = pd.read_csv('G:/My Drive/Ahmad/pyspace/dave/data/dave.csv')


# Plot histograms for each question
for column in df:
    plt.figure()
    sns.histplot(df[column], kde=False, bins=5)
    plt.title(f'Histogram for {column}')
    plt.show()

# Boxplot for each question
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, orient='h', palette="Set2")
plt.title('Boxplots for Each Question')
plt.show()

# Correlation matrix
correlation = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation matrix of Questions')
plt.show()

# Analysis of Correlation matrix
correlation_unstacked = correlation.unstack()
sorted_pairs = correlation_unstacked.sort_values(kind="quicksort")

strong_positive = sorted_pairs[(sorted_pairs > 0.5) & (sorted_pairs < 1)]
strong_negative = sorted_pairs[(sorted_pairs < -0.5)]

print("\n++++++++++++Strongly positively correlated pairs of questions:++++++++\n", strong_positive)
print("\n++++++++++++Strongly negatively correlated pairs of questions:++++++++\n", strong_negative)

```
